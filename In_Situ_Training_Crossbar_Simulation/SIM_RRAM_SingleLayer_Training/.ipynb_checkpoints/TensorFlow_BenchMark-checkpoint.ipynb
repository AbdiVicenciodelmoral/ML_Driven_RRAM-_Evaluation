```python
import numpy as np
import cv2
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
import random
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
def downsample_image(image, size):
    return cv2.resize(image, size, interpolation=cv2.INTER_AREA)

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Define the target size (5x5)
target_size = (10, 10)

# Downsample the training and test images
x_train = np.array([downsample_image(img, target_size) for img in x_train])
x_test= np.array([downsample_image(img, target_size) for img in x_test])

# Verify the downsampling by printing the shape
print("Original size:", x_train.shape)
print("Downsampled size:", x_train.shape)
```

```python
# Display the original and downsampled images

for index in range(0,20):
    plt.subplot(1, 2, 1)
    plt.imshow(x_train[index], cmap='gray')
    plt.title("Original Image (28x28)")

    plt.subplot(1, 2, 2)
    plt.imshow(x_train[index], cmap='gray')
    plt.title("Downsampled Image (5x5)")

    plt.show()
```

```python
# Create a single-layer perceptron model
def create_model(input_size):
    model = Sequential()
    model.add(Dense(1, input_dim=input_size, activation='sigmoid'))
    return model

# Example usage:
input_size = 100  # 10x10 images

# Create a perceptron instance
slp = create_model(input_size)

# Compile the model
slp.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['accuracy'])

# Load your dataset of 10x10 images and corresponding labels
# input_vectors = ...
# labels = ...

# Convert the labels to binary (1 for digit 3, 0 for other digits)
y_train_binary = [1 if label == 3 else 0 for label in y_train]
y_test_binary = [1 if label == 3 else 0 for label in y_test]

# Convert input_vectors and binary_labels into NumPy arrays
input_vectors = np.array(x_train, dtype=np.float32)
binary_labels = np.array(y_train_binary, dtype=np.float32)

# Flatten input_vectors
input_vectors = input_vectors.reshape(-1, 100)

# Train the perceptron
slp.fit(input_vectors, binary_labels, epochs=100, batch_size=32)

# Make predictions for new images
# new_image = ...
# new_image = np.array(new_image, dtype=np.float32).reshape(1, -1)
# is_three = slp.predict(new_image) > 0.5
```

```python
X_test = np.array(x_test, dtype=np.float32)
y_test_binary = np.array(y_test_binary, dtype=np.float32)

# Flatten X_test
X_test = X_test.reshape(-1, 100)

loss, accuracy = slp.evaluate(X_test, y_test_binary)
print("Test set loss:", loss)
print("Test set accuracy:", accuracy)

y_test_pred_prob = slp.predict(X_test)
y_test_pred_binary = (y_test_pred_prob > 0.5).astype(np.int32)
```

```python
import pandas as pd

# Create a dictionary with the data
results_data = {
    "Actual Label": y_test,
    "Predicted Probability": y_test_pred_prob.flatten(),
    "Binary Prediction": y_test_pred_binary.flatten()
}

# Create a pandas DataFrame from the dictionary
results_df = pd.DataFrame(results_data)

# Display the DataFrame
display(results_df)
```
